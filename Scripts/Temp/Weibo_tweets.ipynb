```
import re

from weibo_scraper import  get_formatted_weibo_tweets_by_name
result_iterator = get_formatted_weibo_tweets_by_name(name='皮小蛋的爹', pages=3)
for user_meta in result_iterator:
     for tweetMeta in user_meta.cards_node:
        tstr = tweetMeta.mblog.text 
        x = re.sub('<a .*?><span class="surl-text">', "\n", tstr)
        x = re.sub('</span></a>', " ", x)
        x = re.sub('网页链接',"",x)     #weblinks
        x = re.sub('转发微博',"",x)     #micro-blogging
        x = re.sub('转发确认~',"",x)    #Forward confirmation
        x = re.sub('分享图片',"",x)     #share pictures 
        x = re.sub('哈(哈)+',"",x)      #hahah
        x = re.sub('收',"",x)           #receive
        x = re.sub('Just posted a (photo|video) with Instagram .*',"",x)
        x = re.sub('<br />',"",x)
        x = re.sub('&gt;',"",x)
        x = re.sub(':',"",x)
        x = re.sub('#',"",x)
        x = re.sub("<span .*?>.*</span>","",x)
        x = re.sub("(//)?<a .*?>.*?</a>","",x)
        emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
        x=emoji_pattern.sub(r'', x)# no emoji
        print(x)
        print("----------------------------------------------------")
```
