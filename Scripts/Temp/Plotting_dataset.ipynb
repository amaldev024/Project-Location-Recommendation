```python
import pandas as pd
CheckinInfo=pd.read_csv('checkin_bj_uniq.txt',header=None,sep='\t')
CheckinInfo.columns=['User_id','Time','latitude','longititude','Item_id']
CheckinInfo.head()
```

```python
a1=CheckinInfo.groupby( [ "User_id", "Item_id"] ).size().to_frame(name = 'count').reset_index()
a1.User_id = a1.User_id.astype('int64')
a1.head()
```

```python
import pandas as pd
UserProfile=pd.read_csv('weiboUserProfile_csv.txt',header=None,sep=':')
UserProfile.columns=['User_id','Place','Gender','Year','Horoscope']
UserProfile.User_id = UserProfile.User_id.astype('int64')
UserProfile.head()
```

```python
Checkin_user=pd.merge(a1,UserProfile,on='User_id')
Checkin_user.head()
```

```python
LocationProfile=pd.read_csv('poiInfo_jiepang.txt',header=None,sep=',')
LocationProfile.columns=['Item_id','Latitude','Longitude','Place','Comment_1','Comment_2']
LocationProfile.head()
```

```python
Checkin_user_item=pd.merge(Checkin_user,LocationProfile,on='Item_id')
Checkin_user_item.head()
```

```python
# Data Visualization
#https://cdn-images-1.medium.com/max/800/1*JxbqIQmD_E3M3I7Tjo0OqA.jpeg
# Types:
    # Scatter Plots:width-length-type
    # Line Plots:variation of one
    # Histograms:distribution of data points
    # Bar/Box plots
    # Bar/pie Chart
#http://pythonplot.com 
```

```python
from plotnine import *
import numpy as np
p=(ggplot(Checkin_user_item[(Checkin_user_item.Year > 0) & (Checkin_user_item.Gender != 'nan') & (Checkin_user_item.Year < 100)]) + 
   aes(x='Place_y',y='Year',size='count',color = 'Gender') +
   geom_point() + 
   ggtitle('Scatter Plot') + 
   coord_flip() +
   xlab('Place_y') +
   ylab('Age') 
)
p.save(filename = 'test1.jpeg',height=45,width=10,limitsize=False)
```

```python
Checkin_user_item["Gender"]=Checkin_user_item["Gender"].astype('str')
Checkin_user_item["Horoscope"]=Checkin_user_item["Horoscope"].astype('str')
Checkin_user_item["Comment_1"]=Checkin_user_item["Comment_1"].astype('str')
Checkin_user_item["Comment_2"]=Checkin_user_item["Comment_2"].astype('str')
Checkin_user_item["Place_y"]=Checkin_user_item["Place_y"].astype('str')
Checkin_user_item["Year"]=Checkin_user_item["Year"].astype(float)
Checkin_user_item["Year"]=Checkin_user_item["Year"].fillna(0).astype(int)
Checkin_user_item["Year"]=( 2019 - Checkin_user_item["Year"])
Checkin_user_item["Year"][Checkin_user_item.Year == 2019]=0
```

```python
#Profiling..
import pandas_profiling
pfr = pandas_profiling.ProfileReport(Checkin_user_item)
pfr.to_file("profile.html")
```
