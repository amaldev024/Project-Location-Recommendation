```python
import pandas as pd
CheckinInfo=pd.read_csv('/content/gdrive/My Drive/Dataset/checkin_bj_uniq.txt',header=None,sep='\t')
CheckinInfo.columns=['User_id','Time','latitude','longititude','Item_id']
CheckinInfo.head()
```

```python
a1=CheckinInfo.groupby( [ "User_id", "Item_id"] ).size().to_frame(name = 'count').reset_index()
a1.User_id = a1.User_id.astype('int64')
a1.head()
```

```python
import pandas as pd
UserProfile=pd.read_csv('/content/gdrive/My Drive/Dataset/weiboUserProfile_csv.txt',header=None,sep=':')
UserProfile.columns=['User_id','Place','Gender','Year','Horoscope']
UserProfile.User_id = UserProfile.User_id.astype('int64')
UserProfile.head()
```

```python
Checkin_user=pd.merge(a1,UserProfile,on='User_id')
Checkin_user.head()
```

```python
LocationProfile=pd.read_csv('/content/gdrive/My Drive/Dataset/poiInfo_jiepang.txt',header=None,sep=',')
LocationProfile.columns=['Item_id','Latitude','Longitude','Place','Comment_1','Comment_2']
LocationProfile.head()
```

```python
Checkin_user_item=pd.merge(Checkin_user,LocationProfile,on='Item_id')
Checkin_user_item.head()
```

```python
Checkin_user_item["Gender"]=Checkin_user_item["Gender"].astype('str')
Checkin_user_item["Horoscope"]=Checkin_user_item["Horoscope"].astype('str')
Checkin_user_item["Comment_1"]=Checkin_user_item["Comment_1"].astype('str')
Checkin_user_item["Comment_2"]=Checkin_user_item["Comment_2"].astype('str')
Checkin_user_item["Place_y"]=Checkin_user_item["Place_y"].astype('str')
Checkin_user_item["Year"]=Checkin_user_item["Year"].astype(float)
Checkin_user_item["Year"]=Checkin_user_item["Year"].fillna(0).astype(int)
Checkin_user_item["Year"]=( 2019 - Checkin_user_item["Year"])
Checkin_user_item["Year"][Checkin_user_item.Year == 2019]=0
```

```python
d_item_place_cat = {}
for index,row in Checkin_user_item.iterrows():
  if row.Item_id not in d_item_place_cat.keys():
    d_item_place_cat[row.Item_id]=[row.Place_y,row.Comment_2]
```

```python
#popular place recommendation for new users

user_gender = 'male'
user_age = 30
user_place = 'zunyi'
user_pref = 'theater'
```

```python
#similar gender + age + place + category
print_list = [] 

a=Checkin_user_item[(Checkin_user_item.Gender == user_gender)  & (Checkin_user_item.Year >= user_age - 5) & (Checkin_user_item.Year <= user_age + 5) & (Checkin_user_item.Place_y == user_place) & (Checkin_user_item.Comment_2 == user_pref)]
if (a.shape[0] != 0):
  a = a.sort_values('count',ascending=[False])
  for i in a.Item_id.unique()[0:(a.shape[0] if a.shape[0] < 3 else 3)]:
    print_list.append([str(i),str(d_item_place_cat[i])])  
#similar gender + age + place
a=Checkin_user_item[(Checkin_user_item.Gender == user_gender)  & (Checkin_user_item.Year >= user_age - 5) & (Checkin_user_item.Year <= user_age + 5) & (Checkin_user_item.Place_y == user_place)]
if (a.shape[0] != 0):
  a = a.sort_values('count',ascending=[False])
  for i in a.Item_id.unique()[0:(a.shape[0] if a.shape[0] < 3 else 3)]:
    print_list.append([str(i),str(d_item_place_cat[i])])     
#popular category
abc=Checkin_user_item.groupby(['Item_id','Comment_2'])[["count"]].sum().reset_index()
abc = abc.sort_values('Comment_2')
abc.Comment_2 = abc.Comment_2.astype('str')

temp = abc[abc['Comment_2'] == user_pref]['count'].idxmax()
print_list.append([str(abc.loc[temp][0]),str(d_item_place_cat[abc.loc[temp][0]]),"[most popular place in User_category]"])

#popular place_y
abc=Checkin_user_item.groupby(['Item_id','Place_y'])[["count"]].sum().reset_index()
abc = abc.sort_values('Place_y')
abc.Place_y = abc.Place_y.astype('str')

temp = abc[abc['Place_y'] == user_place]['count'].idxmax()
print_list.append([str(abc.loc[temp][0]),str(d_item_place_cat[abc.loc[temp][0]]),"[most popular place in User_place]"])
```

```python
print('[Popular areas based on preferences]')
col_width = max(len(word) for row in print_list for word in row) + 2  # padding
for row in print_list:
    print("".join(word.ljust(col_width) for word in row))
```
