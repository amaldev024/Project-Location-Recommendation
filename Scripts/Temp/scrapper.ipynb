```
#install and import packages
!pip install bs4
import requests
from bs4 import BeautifulSoup
import csv

#adding page number to pages. "crawl" the first 100 pages starting from page 2
for i in range(2,100):
    link_old="https://www.themobileindian.com/product-finder/search/mobile/page-"
    #new link of the website page
    link=link_old+str(i)    
    response = requests.get(link)
    soup=BeautifulSoup(response.text,"html.parser")
    
    #class containing all gadgets information
    gadgets=soup.find_all(class_="gadget-card")

    #writing all text to input2.txt(contains space and unnecessary lines)
    with open("input2.txt","w") as f:
        for gadget in gadgets:
            f.write(gadget.get_text())

    #removing blank lines
    with open("input2.txt","r") as f:
        with open('input1.txt','w') as s:
            for line in f:
                if not line.isspace():
                    s.write(line)

    #removing lines "Add to compare" (unnecessary lines)
    with open('input1.txt','r') as f:
        with open('input.txt','w') as s:
            for line in f:
                if(line==" Add to compare\n"):
                    continue
                else:
                    s.write(line)
                    
    #creating dataset
    counter=0
    with open("input.txt","r") as f:
        with open("output.txt","a") as s:
            for line in f:
                counter+=1;
                line2=line.strip()
                s1,*args=line2.split()
                if(s1=="Product" or s1=="See" or s1=="Price"):
                    continue
                elif(s1=="Operating"):
                    s2,s3=line2.split(":")
                    operating=s3.strip()
                elif(s1=="Processor"):
                    s2,s3=line2.split(":")
                    processor=s3.strip()
                elif(s1=="Battery"):
                    s2,s3=line2.split(":")
                    battery=s3.strip()
                elif(s1=="Resolution"):
                    s2,s3=line2.split(":")
                    resolution=s3.strip()
                elif(s1=="RAM"):
                    s2,s3=line2.split(":")
                    ram=s3.strip()
                elif(s1=="Screen"):
                    s2,s3=line2.split(":")
                    screen=s3.strip()
                elif(s1=="Internal"):
                    s2,s3=line2.split(":")
                    internal=s3.strip()
                elif(s1=="Front"):
                    s2,s3=line2.split(":")
                    front=s3.strip()
                elif(s1=="Rear"):
                    s2,s3=line2.split(":")
                    rear=s3.strip()
                elif(s1=="Expandable"):
                    s2,s3=line2.split(":")
                    expandable=s3.strip()
                elif(s1=="Rs."):
                    s2,s3,*args=line2.split(" ")
                    rs=s3
                else:
                    device=line2  
                #each new device start in the 20th line.
                if(counter==20):
                    s.write(device+":"+operating+":"+processor+":"+battery+":"+resolution+":"+ram+":"+screen+":"+internal+":"+front+":"+rear+":"+expandable+":"+rs)
                    s.write("\n")
                    counter=0
```
