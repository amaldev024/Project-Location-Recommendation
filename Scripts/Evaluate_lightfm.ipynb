```
#generation user - item matrix
from scipy.sparse import csr_matrix
from pandas.api.types import CategoricalDtype

a1=CheckinInfo.groupby( [ "User_id", "Item_id"] ).size().to_frame(name = 'count').reset_index()
#a1.loc[a1['User_id'].isin(new_user_id),'count'] = 0

person_c = CategoricalDtype(sorted(a1.User_id.unique()), ordered=True)
thing_c = CategoricalDtype(sorted(a1.Item_id.unique()), ordered=True)

items=sorted(a1["Item_id"].unique())
user_labels=list(sorted(a1.User_id.unique()))

#Converting sparse-matrix to dataframe 
#dfs = pd.SparseDataFrame(sparse_checkin, index=person_c.categories, columns=thing_c.categories,default_fill_value=0)
```

```
#generating user-feature and item-feature matrices
from lightfm.data import Dataset
dataset = Dataset()

item_labels=sorted(a1["Item_id"].unique())
user_labels=list(sorted(a1.User_id.unique()))

temp_pd = UserProfile[UserProfile.User_id.isin(user_labels)].sort_values(['User_id'], ascending=[True]).reset_index(drop=True)                                                                   
u_features = list()
u_features.append(list(temp_pd.Place.unique()))
u_features.append(list(temp_pd.Gender.unique()))
u_features.append(list(temp_pd.Year.unique()))
u_features.append(list(temp_pd.Horoscope.unique())) 

def add_to_list(tweet_str):
    u_features.append(tweet_str.split(';'))
    return tweet_str
temp_pd['tweet'][:].apply(add_to_list)
u_features = [j for i in u_features for j in i]
u_features = list(dict.fromkeys(u_features))

temp_pd2 = LocationProfile[LocationProfile.Item_id.isin(item_labels)].sort_values(['Item_id'], ascending=[True]).reset_index(drop=True)                                                                    
i_features = list()
i_features.append(list(temp_pd2.Place.unique()))
i_features.append(list(temp_pd2.Comment_1.unique()))
i_features.append(list(temp_pd2.Comment_2.unique()))
i_features = [j for i in i_features for j in i]



dataset.fit(user_labels,item_labels,user_features = u_features,item_features = i_features)
```

```
user_features_matrix = dataset.build_user_features(((x['User_id'], [x['Place'],x['Gender'],x['Year'],x['Horoscope']] + x.tweet.split(';'))
                                                    for index,x in temp_pd.iterrows()))

item_features_matrix = dataset.build_item_features(((x['Item_id'], [x['Place'],x['Comment_1'],x['Comment_2']])
                                                    for index,x in temp_pd2.iterrows()))
```

```
# hitrate - one item hide
import random
d1={} #all user-item
d2={} #hidden user-item

User_place_count=a1.groupby( [ "User_id"] ).size().to_frame(name = 'count').reset_index()
User_place_count.count = User_place_count["count"].astype('int64')
User_place_count.User_id = User_place_count["User_id"].astype('int64')

#generating single user dictionary
dict_singleUser={}
for index,row in User_place_count[:][User_place_count["count"] == 1].iterrows():
    dict_singleUser[row.User_id] = 0;
    
#generating d[user]=item(s)
for index,row in a1.iterrows():
    key1=row.User_id
    key2=row.Item_id
    if key1 not in dict_singleUser.keys():
        if key1 not in d1.keys():
            d1[key1]=[key2]
        else:
            d1[key1].append(key2)
```

```
# hitrate - one item hide
import tqdm
list_toBeDeleted=[]
visited=[]

with tqdm.tqdm(total=a1.User_id.count()) as progress:
  for index,row in a1.iterrows():
      if row.User_id not in visited and row.User_id not in dict_singleUser:
          temp=random.choice(d1[row.User_id])
          d2[row.User_id]=temp
          d1[row.User_id].remove(temp)
          visited.append(row.User_id)
          list_toBeDeleted.append(index)
      progress.update(1)
        
```

```
# hitrate - one item hide
a1.drop(list_toBeDeleted, inplace=True)
rows = a1.User_id.astype(person_c).cat.codes
cols = a1.Item_id.astype(thing_c).cat.codes
sparse_checkin = csr_matrix((a1['count'], (rows, cols)),shape=(person_c.categories.size, thing_c.categories.size))
```

```
from lightfm import LightFM
from lightfm.evaluation import precision_at_k

# Instantiate and train the model
model = LightFM(loss='warp',no_components=50)

import time
print("training model :")
start = time.time()
model.fit(sparse_checkin, epochs=20, num_threads=40)#,item_features=item_features_matrix,user_features=user_features_matrix)
print("trained model in %0.2fs" %(time.time() - start))



```
